```python
import json
```

```python
input_address = '2-StrategyQA\\strategyqa_train.json'
output_address = '2-StrategyQA\\StrategyQA.json'

instruction = 'Answer this Question: '

samples_list = []

with open(input_address, 'r', encoding="utf8") as openfile:
    json_object = json.load(openfile)
    
    for train_sample in json_object:
        print(train_sample)
        
        problem = train_sample["question"]
        print('----- Problem:', problem)
        answer = train_sample["answer"]
        print('----- Answer:', answer)
        cot = ''
        facts = train_sample["facts"]
        for fact in facts:
            cot += fact + ' '
        print('----- COT:', cot)
        
        sample = {
            "instruction" : instruction,
            "problem" : problem,
            "cot" : cot,
            "answer" : answer
        }
        samples_list.append(sample)
        
        print('*'*100)
        

json_object = json.dumps(samples_list, indent = 4)
with open(output_address, "w") as output_file:
    output_file.write(json_object)
```
